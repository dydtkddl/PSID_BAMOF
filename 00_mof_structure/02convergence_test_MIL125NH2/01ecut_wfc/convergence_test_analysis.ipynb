```python
import os 
dirs = [ x for x in  os.listdir() if os.path.isdir(x)]
dc ={}
for i in dirs:
    pa = os.path.join(i , "input.out")
    with open(pa, "r") as f:
        a = f.readlines()
    for aa in a:
        if aa[0] == "!":
            dc[pa] = float(aa.replace("!    total energy              =   ", "").replace(" Ry\n", ""))
            break 
dc
import matplotlib.pyplot as plt 
plt.plot( [30,40,50,60] , [ x for x in dc.values()])
# plt.scatter( [30,40,50,60] , dc.values(), marker = "*", color = "red")
0.3 * 2600
```
