```python
import os 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
SAT_PRESSURE = 1.013e5

df = pd.read_csv("./07result.csv")
df = df[["name", "Average loading absolute [mol/kg framework]", "gas"	,"pressure"	,"temp"]]
# df
df["pressure"] = 100000*df["pressure"]
df["rel_pressure"] = df["pressure"] /SAT_PRESSURE
```

```python
df
```

```python
import matplotlib.pyplot as plt

# BET    
df["v"] = df["Average loading absolute [mol/kg framework]"]
df["x"] = df["rel_pressure"]
df["y"] = 1 / (df["v"] * (1/df["x"] - 1))

# BET plot  ( )
fig, ax = plt.subplots()

for name, group in df.groupby("name"):
    ax.plot(group["x"], group["y"], 'o-', label=name)

ax.set_xlabel("Relative Pressure (P/P)")
ax.set_ylabel(r"1 / [v (P/P - 1)]")
ax.set_title("BET Plot")
ax.legend()
ax.grid(True)

plt.show()
```
